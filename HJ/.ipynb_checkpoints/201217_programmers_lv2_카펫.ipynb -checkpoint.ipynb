```python
# 가로 >=  세로
def solution(brown, yellow):
    answer = []
    for x in range(8,5000):
        for y in range(1,2000000):
            if x==y:print(x,y)
            elif x<y:break
            elif 2*(x+y)-4==brown and (x-2)*(y-2)==yellow:
                return[x,y]
    
    return answer
solution(10,2)
```

```python
##개같은 수학자놈들!
def solution(brown, red):
    for i in range(1, int(red**(1/2))+1):
        if red % i == 0:
            if 2*(i + red//i) == brown-4:
                return [red//i+2, i+2]
```
